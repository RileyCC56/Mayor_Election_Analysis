```python
# Dependencies
import numpy as np
import pandas as pd
import requests

# API Key
from config import apikey
```

```python
# Read the data
cont_data_df = pd.read_csv("./Resources/Clean_Committee_2017.csv", low_memory=False)
cont_data_df.head()
```

```python
# Drop unnecessary columns
cont_data_df = cont_data_df.drop(['Election', 'CandidateID', 'Date', 'ContributionState', 'ContributerType', 'MatchAmt', 'PayMethod'], axis=1)
cont_data_df
```

```python
# Regularize numerical columns
cont_data_df['ZipCode'] = cont_data_df['ZipCode'].astype(int)
cont_data_df['Amount'] = cont_data_df['Amount'].astype(int)
cont_data_df['PrevAmt'] = cont_data_df['PrevAmt'].astype(int)
cont_data_df['TotalAmount'] = cont_data_df['TotalAmount'].astype(int)
cont_data_df
```

```python
# Export cleaned data
cont_data_df.to_csv("./Resources/Clean_Committee_2017_Revised.csv", encoding='utf-8', index=False)
```

```python
# Read the data for the next file
cont_data2_df = pd.read_csv("./Resources/Clean_Individual_2017.csv", low_memory=False)
cont_data2_df.head()
```

```python
# Drop unnecessary columns
cont_data2_df = cont_data2_df.drop(['Election', 'CandidateID', 'Date', 'ContributionState', 'ContributerType', 'MatchAmt', 'PayMethod'], axis=1)
cont_data2_df
```

```python
# Regularize numerical columns
cont_data2_df['ZipCode'] = cont_data2_df['ZipCode'].dropna()
cont_data2_df['ZipCode'] = cont_data2_df['ZipCode'].round(2)
cont_data2_df['Amount'] = cont_data2_df['Amount'].round(2)
cont_data2_df['PrevAmt'] = cont_data2_df['PrevAmt'].round(2)
cont_data2_df['TotalAmount'] = cont_data2_df['TotalAmount'].round(2)
cont_data2_df
```

```python
# Export cleaned data
cont_data_df.to_csv("./Resources/Clean_Individual_2017_Revised.csv", encoding='utf-8', index=False)
```
