```python
# Dependencies
import numpy as np
import pandas as pd
import requests
```

```python
# Read the data
cont_data_df = pd.read_csv("./Resources/Clean_Committee_2017.csv", low_memory=False)
cont_data_df.head()
```

```python
# Drop unnecessary columns
cont_data_df = cont_data_df.drop(['CandidateID', 'Date', 'ContributionState', 'ContributerType', 'MatchAmt', 'PayMethod'], axis=1)
cont_data_df
```

```python
# Drop all blank columns
cont_data_df = cont_data_df.dropna()
cont_data_df
```

```python
# Regularize numerical columns
cont_data_df['ZipCode'] = cont_data_df['ZipCode'].astype(int)
cont_data_df['Amount'] = cont_data_df['Amount'].astype(int)
cont_data_df['PrevAmt'] = cont_data_df['PrevAmt'].astype(int)
cont_data_df['TotalAmount'] = cont_data_df['TotalAmount'].astype(int)
cont_data_df
```

```python
# Converting dollar amounts according to inflation (figures derived from https://www.usinflationcalculator.com/)
cont_data_df['ElectionTemp'] = cont_data_df['Election']
cont_data_df.loc[cont_data_df['ElectionTemp'] == 2001, 'AdjustedAmount'] = cont_data_df['Amount']*1.55
cont_data_df['Election'] = cont_data_df['ElectionTemp']
cont_data_df.loc[cont_data_df['ElectionTemp'] == 2005, 'AdjustedAmount'] = cont_data_df['Amount']*1.40
cont_data_df['Election'] = cont_data_df['ElectionTemp']
cont_data_df.loc[cont_data_df['ElectionTemp'] == 2009, 'AdjustedAmount'] = cont_data_df['Amount']*1.28
cont_data_df['Election'] = cont_data_df['ElectionTemp']
cont_data_df.loc[cont_data_df['ElectionTemp'] == 2013, 'AdjustedAmount'] = cont_data_df['Amount']*1.17
cont_data_df['Election'] = cont_data_df['ElectionTemp']
cont_data_df.loc[cont_data_df['ElectionTemp'] == 2017, 'AdjustedAmount'] = cont_data_df['Amount']*1.11
cont_data_df['Election'] = cont_data_df['ElectionTemp']
cont_data_df.loc[cont_data_df['ElectionTemp'] == 2021, 'AdjustedAmount'] = cont_data_df['Amount']*1.00

cont_data_df['ElectionTemp'] = cont_data_df['Election']
cont_data_df.loc[cont_data_df['ElectionTemp'] == 2001, 'AdjustedPrevAmt'] = cont_data_df['PrevAmt']*1.55
cont_data_df['Election'] = cont_data_df['ElectionTemp']
cont_data_df.loc[cont_data_df['ElectionTemp'] == 2005, 'AdjustedPrevAmt'] = cont_data_df['PrevAmt']*1.40
cont_data_df['Election'] = cont_data_df['ElectionTemp']
cont_data_df.loc[cont_data_df['ElectionTemp'] == 2009, 'AdjustedPrevAmt'] = cont_data_df['PrevAmt']*1.28
cont_data_df['Election'] = cont_data_df['ElectionTemp']
cont_data_df.loc[cont_data_df['ElectionTemp'] == 2013, 'AdjustedPrevAmt'] = cont_data_df['PrevAmt']*1.17
cont_data_df['Election'] = cont_data_df['ElectionTemp']
cont_data_df.loc[cont_data_df['ElectionTemp'] == 2017, 'AdjustedPrevAmt'] = cont_data_df['PrevAmt']*1.11
cont_data_df['Election'] = cont_data_df['ElectionTemp']
cont_data_df.loc[cont_data_df['ElectionTemp'] == 2021, 'AdjustedPrevAmt'] = cont_data_df['PrevAmt']*1.00

cont_data_df['ElectionTemp'] = cont_data_df['Election']
cont_data_df.loc[cont_data_df['ElectionTemp'] == 2001, 'AdjustedTotalAmount'] = cont_data_df['TotalAmount']*1.55
cont_data_df['Election'] = cont_data_df['ElectionTemp']
cont_data_df.loc[cont_data_df['ElectionTemp'] == 2005, 'AdjustedTotalAmount'] = cont_data_df['TotalAmount']*1.40
cont_data_df['Election'] = cont_data_df['ElectionTemp']
cont_data_df.loc[cont_data_df['ElectionTemp'] == 2009, 'AdjustedTotalAmount'] = cont_data_df['TotalAmount']*1.28
cont_data_df['Election'] = cont_data_df['ElectionTemp']
cont_data_df.loc[cont_data_df['ElectionTemp'] == 2013, 'AdjustedTotalAmount'] = cont_data_df['TotalAmount']*1.17
cont_data_df['Election'] = cont_data_df['ElectionTemp']
cont_data_df.loc[cont_data_df['ElectionTemp'] == 2017, 'AdjustedTotalAmount'] = cont_data_df['TotalAmount']*1.11
cont_data_df['Election'] = cont_data_df['ElectionTemp']
cont_data_df.loc[cont_data_df['ElectionTemp'] == 2021, 'AdjustedTotalAmount'] = cont_data_df['TotalAmount']*1.00
cont_data_df
```

```python
# Export cleaned data
cont_data_df.to_csv("./Resources/Clean_Committee_2017.csv", encoding='utf-8', index=False)
```
